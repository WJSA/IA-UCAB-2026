```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar los datos
df = pd.read_csv('datosLabRegresion/datosVentas.csv')

X = df[['Año']].values
y = df['Ventas'].values

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Predecir para el año 2026
X_pred = np.array([[2026]])
prediccion_2026 = modelo.predict(X_pred)

print(f"Predicción de ventas para el año 2026: {prediccion_2026[0]:.2f}")
print(f"Coeficiente del modelo: {modelo.coef_[0]:.4f}")
print(f"Intercepto del modelo: {modelo.intercept_:.4f}")

# Gráfica
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), color='red', label='Ajuste lineal')
plt.scatter(X_pred, prediccion_2026, color='green', marker='X', s=100, label='Predicción 2026')
plt.xlabel('Año')
plt.ylabel('Ventas')
plt.title('Regresión Lineal: Predicción de Ventas')
plt.legend()
plt.grid(True)
plt.show()
```
