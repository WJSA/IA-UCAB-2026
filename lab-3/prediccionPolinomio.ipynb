```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Cargar datos
df = pd.read_csv('datosLabRegresion/datosProblema2.csv')
X = df[['X']].values
y = df['y'].values

# Usaremos un polinomio de grado 3 (se puede ajustar si se requiere)
grado = 3
poly = PolynomialFeatures(degree=grado)
X_poly = poly.fit_transform(X)

modelo = LinearRegression()
modelo.fit(X_poly, y)

# Puntos a predecir
X_pred = np.array([[0], [1.5], [3], [5]])
X_pred_poly = poly.transform(X_pred)
y_pred = modelo.predict(X_pred_poly)

print("Predicciones:")
for val_x, val_y in zip(X_pred.flatten(), y_pred):
    print(f"X = {val_x} -> y_pred = {val_y:.4f}")

# Gráfica
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Datos reales', alpha=0.5)

# Crear curva suave para la gráfica del modelo polinómico
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly.transform(X_plot)
plt.plot(X_plot, modelo.predict(X_plot_poly), color='red', label=f'Ajuste polinómico (grado {grado})')

# Graficar los puntos predichos
plt.scatter(X_pred, y_pred, color='green', marker='X', s=100, label='Predicciones solicitadas')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Polinómica')
plt.legend()
plt.grid(True)
plt.show()
```
