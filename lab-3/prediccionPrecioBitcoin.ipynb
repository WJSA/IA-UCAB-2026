```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import datetime

# 1. Cargar y preparar los datos
df = pd.read_csv('datosLabRegresion/datosBitcoin.csv')

# Limpiar los precios (remover comas y convertir a float)
df['Price'] = df['Price'].astype(str).str.replace(',', '').astype(float)

# Convertir la columna Date a tipo datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Para la regresión, convertiremos las fechas a su valor ordinal
df['Date_ordinal'] = df['Date'].apply(lambda x: x.toordinal())

# Ordenar los datos cronológicamente (opcional pero recomendado para graficar)
df = df.sort_values('Date')

X = df[['Date_ordinal']].values
y = df['Price'].values

# 2. Entrenar modelo
modelo = LinearRegression()
modelo.fit(X, y)

# 3. Predicciones solicitadas
fechas_pred_str = ['04/23/2026', '04/24/2026', '04/25/2026']
fechas_pred_dt = pd.to_datetime(fechas_pred_str, format='%m/%d/%Y')
X_pred = np.array([d.toordinal() for d in fechas_pred_dt]).reshape(-1, 1)

precios_predichos = modelo.predict(X_pred)

# NOTA: Los "precios reales" deben ser consultados o proporcionados manualmente
# Aquí colocamos unos precios aproximados. Reemplázalos por los precios reales de cierre de esos días.
precios_reales = [78257.48, 77437.13, 77625.00] # <-- Valores reales consultados para el 23, 24 y 25 de abril

print("Resultados de la Predicción del Bitcoin:")
print("-" * 50)
for i in range(len(fechas_pred_str)):
    val_real = precios_reales[i]
    val_pred = precios_predichos[i]
    porcentaje_dv = abs(val_real - val_pred) / val_real * 100
    
    print(f"Fecha: {fechas_pred_str[i]}")
    print(f"  Precio real     : ${val_real:,.2f}")
    print(f"  Precio predicho : ${val_pred:,.2f}")
    print(f"  % Desviación    : {porcentaje_dv:.2f}%\n")

# 4. Gráfica
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], y, color='blue', label='Precio Histórico (Real)', alpha=0.7)

# Línea de regresión sobre los datos históricos
y_ajuste = modelo.predict(X)
plt.plot(df['Date'], y_ajuste, color='red', label='Ajuste de Regresión Lineal')

# Puntos predichos
plt.scatter(fechas_pred_dt, precios_predichos, color='green', marker='X', s=100, label='Predicciones (23, 24, 25 Abr)')

plt.xlabel('Fecha')
plt.ylabel('Precio Bitcoin (USD)')
plt.title('Regresión: Predicción del Precio del Bitcoin')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
